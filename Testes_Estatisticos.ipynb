```python
zconfint(notas1.rating)
```

```python
from statsmodels.stats.weightstats import ztest

ztest(notas1.rating, value = 3.4320503405352594)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac = 1).rating

def calcula_teste(i):
    medias  = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value = 3.4320503405352594)
    return i, medias, p
    
valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:, 0], valores[:,1], label = "Média")
plt.plot(valores[:, 0], valores[:, 2], label = 'Ztest')
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), color = 'k')
plt.xlabel("Amostras")
plt.title("Comparação de Número de Amostras X Média / Ztest")
plt.legend()
```

<h1/>Testando se é Distribuição Normal<h1/>

```python
from scipy.stats import normaltest
_, p = normaltest(notas1.rating)
p
```

<h3/>A distribuição não é normal, logo é preciso se aplicar um teste não-paramétrico<h3/>

```python
from scipy.stats import ranksums

_, p = ranksums(notas1.rating, notas.rating)
p
```
